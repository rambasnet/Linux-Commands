## File System Orgranization

- the first directory in the file system is called *root directory*
- root directory / contains files and subdirectories which contain more files and subdirectories and so on

## file name and paths
- file names are case sensitive
- do NOT use spaces in file and folder names
    - navigation hashle as you've to escape spaces with \
- file paths can be:
    - absolute: starting from the root /
    - relative: starting from the working directory
- shortcuts:
    - **~** : home direcotry of current loggedin user
    - **~/user/** : home directory of user
    - **cd** : cd without filder name changes directory to home dir
    - **cd -**: changes directory to previous one
